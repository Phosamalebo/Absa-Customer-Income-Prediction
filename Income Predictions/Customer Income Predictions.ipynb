```python
Income_group = Income_group.rename(columns={'income_group_code': 'INCOME_GROUP_CODE', 'income_group_desc': 'INCOME_GROUP_DESC'})
```

```python
transactions = transactions.rename(columns={'RECORD_DATE': 'TRANSACTION_RECORD_DATE'})
```

```python
Train = Train.rename(columns={'RECORD_DATE': 'TRAIN_RECORD_DATE'})
```

```python
df1 = pd.merge(customer,                 
                      transactions,
                      on = "CUSTOMER_IDENTIFIER",
                      how = "inner")

df1
```

```python
df1.info()
```

```python
df1.head()
```

```python
df = pd.merge(df1,                 
                      Train,
                      on = "CUSTOMER_IDENTIFIER",
                      how = "inner")

df.head()
```

```python
df_merged = pd.merge(df, Income_group, on = "INCOME_GROUP_CODE", how = "inner")
df_merged.head()
```

```python
df_merged.info()
```

```python
df_merged = df_merged.drop('TRANSACTION_REFERENCE', axis=1)
```

```python
df_merged.head()
```

```python
df_merged['TRANSACTION_DESCRIPTION'].fillna('NO DESCRIPTION', inplace=True)
df_merged.head()
```

```python
df_merged['TRANSACTION_RECORD_DATE'] = pd.to_datetime(df_merged['TRANSACTION_RECORD_DATE'])
df_merged['MONTH'] = pd.to_datetime(df_merged['TRANSACTION_RECORD_DATE']).dt.month_name()

df_merged.head()
```

```python
df_merged.isna().sum()
```

```python
df_merged.to_csv('cleaned_df.csv')
print('The CSV file has been exported successfully.')
```
